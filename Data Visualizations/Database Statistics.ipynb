```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# This dataset is the learning dataset
master_dataset = pd.read_csv('Database_ver2.csv')
master_dataset['Coercivity'] = np.log(master_dataset['Coercivity'])
master_dataset.rename(columns={'Coercivity':'Log(Coercivity)'},inplace=True)
master_dataset.head()
```

```python
SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 22

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
```

```python
import warnings
warnings.filterwarnings("ignore")
fig, ax = plt.subplots(figsize=(5, 5))

plt.title("Magnetic Saturation Distribution")
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax.set_ylim(bottom=0,top=2)
ax.set_xlim(left=0,right=2)
mag_sat = sns.distplot(master_dataset['Magnetic Saturation'].dropna(), bins = 15, kde_kws={"color": "#f06449", "label": "KDE","alpha":1},hist_kws={"color": "#0a2364","alpha":1})
plt.xlabel("Magnetic Saturation (T)")
plt.ylabel("Density")
mag_sat.figure.savefig('Figures\\CuratedDB\\Distributions\\magnetic_saturation_poster.png', bbox_inches='tight',dpi=300)
```

```python
import warnings
warnings.filterwarnings("ignore")
fig, ax = plt.subplots(figsize=(5, 5))

plt.title("Magnetostriction Distribution")
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.02))
ax.set_ylim(bottom=0,top=0.1)
ax.set_xlim(left=-10,right=35)
magneto = sns.distplot(master_dataset['Magnetostriction'].dropna(),bins = 10,kde_kws={"color": "#f06449", "label": "KDE","alpha":1},hist_kws={"color": "#0a2364","alpha":1})
plt.xlabel("Magnetostriction (10^-6)")
plt.ylabel("Density")
magneto.figure.savefig('Figures\\CuratedDB\\Distributions\\magnetostriction_poster.png', bbox_inches='tight',dpi=300)
```

```python
import warnings
warnings.filterwarnings("ignore")
fig, ax = plt.subplots(figsize=(5, 5))

plt.title("Log(Coercivity) Distribution")
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.05))
ax.set_ylim(bottom=0,top=0.3)
ax.set_xlim(left=-4,right=10)
coercivity = sns.distplot(master_dataset['Log(Coercivity)'].dropna(),bins = 20,kde_kws={"color": "#f06449", "label": "KDE","alpha":1},hist_kws={"color": "#0a2364","alpha":1})
plt.xlabel("Log(Coercivity(A/m))")
plt.ylabel("Density")
coercivity.figure.savefig('Figures\\CuratedDB\\Distributions\\coercivity_poster.png', bbox_inches='tight',dpi=300)
```
