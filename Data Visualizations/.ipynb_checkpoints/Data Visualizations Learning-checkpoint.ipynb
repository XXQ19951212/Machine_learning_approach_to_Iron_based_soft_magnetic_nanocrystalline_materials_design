```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This dataset is the learning dataset
master_dataset = pd.read_csv('LearningDatabase7-26.csv')
master_dataset.head()
```

```python
# Transforming variables with skewed distributions (Coercivity and Permeability)
master_dataset['Coercivity'] = np.log(master_dataset['Coercivity'])
master_dataset['Permeability'] = np.log(master_dataset['Permeability'])
master_dataset.rename(columns={'Coercivity':'Log(Coercivity)','Permeability':'Log(Permeability)'},inplace=True)
master_dataset
```

```python
master_dataset['Reference DOI'].nunique()
```

```python
# Selecting composition fingerprints
composition = list(master_dataset.iloc[:,28:53])
new_composition = list(master_dataset.iloc[:,:25])
```

```python
# Group dataset with regards to fixed compositions
dataset_by_composition = master_dataset.groupby(composition)
```

```python
# Create dataframe with group and counts
comp_ids = []
comp_counts = []
for name,group in dataset_by_composition:
    comp_ids.append(name)
    comp_counts.append(len(group))

comp_instances= list(zip(comp_ids,comp_counts))

comp_data = pd.DataFrame(comp_instances,columns=['ID','Count'])
comp_data
```

```python
# Selecting all of the original features
old_features = list(master_dataset.iloc[:,28:61])
```

```python
# Selecting all of the engineered features
new_features = list(master_dataset.iloc[:,0:28])
```

```python
# Selecting all of the property values 
values = list(master_dataset.iloc[:,61:69])
```

```python
# Selecting all of the meta features
meta = ['Reference DOI','Composition ID']
```

```python
# Plots showing the scatter between the the two sets of features and values
old_scatter = sns.PairGrid(master_dataset, x_vars=old_features,y_vars=values).map(plt.scatter, color='#016FB9')
new_scatter = sns.PairGrid(master_dataset, x_vars=new_features,y_vars=values).map(plt.scatter, color='#016FB9')
```

```python
# Plots showing the scatter (with regression lines) between the the two sets of features and values
old_regplot = sns.PairGrid(master_dataset, x_vars=old_features,y_vars=values).map(sns.regplot,ci=None)
new_regplot = sns.PairGrid(master_dataset, x_vars=new_features,y_vars=values).map(sns.regplot,ci=None)
```

```python
# Save all the figures to disk
old_scatter.savefig('Figures\\CuratedDB\\c_old_scatter.png')
new_scatter.savefig('Figures\\CuratedDB\\c_new_scatter.png')
old_regplot.savefig('Figures\\CuratedDB\\c_old_regplot.png')
new_regplot.savefig('Figures\\CuratedDB\\c_new_regplot.png')
```

```python
# Plot showing the scatter between values
value_scatter = sns.PairGrid(master_dataset[values])
value_scatter.map_lower(plt.scatter,color='darkblue')
for i, j in zip(*np.triu_indices_from(value_scatter.axes, 1)):
    value_scatter.axes[i, j].set_visible(False)
for i in range(len(values)):
    value_scatter.axes[i,i].set_visible(False)
```

```python
# Plot showing the regression between values
value_regplot = sns.PairGrid(master_dataset[values])
value_regplot.map_lower(sns.regplot, ci=None)
for i, j in zip(*np.triu_indices_from(value_regplot.axes, 1)):
    value_regplot.axes[i, j].set_visible(False)
for i in range(len(values)):
    value_regplot.axes[i,i].set_visible(False)
```

```python
# Outputting plots
value_scatter.savefig('Figures\\CuratedDB\\c_value_scatter.png')
value_regplot.savefig('Figures\\CuratedDB\\c_value_regplot.png')
```

```python
# Generate the correlation matrix for the entire dataset
dataset_correlations = master_dataset.drop(columns=meta).corr()
```

```python
# Extract only the correlations between the values and the old features
old_feature_correlations = dataset_correlations[61:][old_features]
```

```python
# Extract only the correlations between the values and the new features
new_feature_correlations = dataset_correlations[61:][new_features]
```

```python
# Extract only the correlations between the values
value_correlations = dataset_correlations[61:][values]
```

```python
# Define a color palette
cmap = sns.diverging_palette(220, 10, as_cmap=True)
```

```python
# Plot the old feature correlations as a heatmap 
f_old, ax_old = plt.subplots(figsize=(25,5))
old_corr_heatmap = sns.heatmap(old_feature_correlations, cmap = cmap, linewidths=0.5, square = True)
```

```python
# Plot the new feature correlations as a heatmap 
f_new, ax_new = plt.subplots(figsize=(20,5))
new_corr_heatmap = sns.heatmap(new_feature_correlations, cmap = cmap, linewidths=0.5, square = True)
```

```python
# Exporting feature heatmaps
old_corr_heatmap.figure.savefig('Figures\\CuratedDB\\c_old_features_correlations_heatmap.png', bbox_inches='tight')
new_corr_heatmap.figure.savefig('Figures\\CuratedDB\\c_new_features_correlations_heatmap.png', bbox_inches='tight')
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(value_correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Plot the value correlations as a heatmap
f_val, ax_val = plt.subplots(figsize=(8,5))
val_corr_heatmap = sns.heatmap(value_correlations, mask = mask, cmap = cmap, linewidths=0.5, square = True)
```

```python
# Exporting value heatmap 
val_corr_heatmap.figure.savefig('Figures\\CuratedDB\\c_value_correlations_heatmap.png', bbox_inches='tight')
```

```python
# Generating histograms of feature values to visualize distributions
import warnings
warnings.filterwarnings("ignore")

distribution={}
for column in list(master_dataset.drop(columns=meta)):
    ax = plt.plot(figsize=(5, 5))
    distribution[column] = sns.distplot(master_dataset.drop(columns=meta)[column].dropna())
    plt.close()
```

```python
# Exporting distributions as figures
for i in distribution:
    distribution[i].figure.savefig('Figures\\CuratedDB\\Distributions\\' + i +'.png', bbox_inches='tight')
```
